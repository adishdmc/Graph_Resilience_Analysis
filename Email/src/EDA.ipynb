 Importing Libraries

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import json
```

Load the Preprocessed Email Graph and Labels

```python
email_graph = nx.read_gml('data/email_graph.gml')
with open('data/email_labels.json', 'r') as f:
    email_labels = json.load(f)
```

 Basic Information and Statistics

```python
print("Number of nodes:", email_graph.number_of_nodes())
print("Number of edges:", email_graph.number_of_edges())
print("Average degree:", sum(dict(email_graph.degree()).values()) / email_graph.number_of_nodes())
print("Density:", nx.density(email_graph))
print("Average clustering coefficient:", nx.average_clustering(email_graph))
```

Degree Distribution

```python
degrees = [degree for node, degree in email_graph.degree()]
plt.figure(figsize=(10, 6))
plt.hist(degrees, bins=20, edgecolor='black')
plt.xlabel('Degree')
plt.ylabel('Frequency')
plt.title('Degree Distribution (Email Dataset)')
plt.show()
```

Clustering Coefficient Distribution

```python
clustering_coeffs = list(nx.clustering(email_graph).values())
plt.figure(figsize=(10, 6))
plt.hist(clustering_coeffs, bins=20, edgecolor='black')
plt.xlabel('Clustering Coefficient')
plt.ylabel('Frequency')
plt.title('Clustering Coefficient Distribution (Email Dataset)')
plt.show()
```

 Degree vs. Clustering Coefficient Scatterplot

```python
plt.figure(figsize=(10, 6))
plt.scatter(degrees, clustering_coeffs, alpha=0.6)
plt.xlabel('Degree')
plt.ylabel('Clustering Coefficient')
plt.title('Degree vs. Clustering Coefficient (Email Dataset)')
plt.show()
```

Betweenness Centrality Distribution

```python
betweenness = nx.betweenness_centrality(email_graph)
betweenness_values = list(betweenness.values())
plt.figure(figsize=(10, 6))
plt.hist(betweenness_values, bins=20, edgecolor='black')
plt.xlabel('Betweenness Centrality')
plt.ylabel('Frequency')
plt.title('Betweenness Centrality Distribution (Email Dataset)')
plt.show()
```

Visualization of the Network

```python
plt.figure(figsize=(12, 12))
pos = nx.spring_layout(email_graph, seed=42)
nx.draw(email_graph, pos, with_labels=False, node_size=20, edge_color='k', alpha=0.5)
plt.title('Email Network Visualization')
plt.show()
```
