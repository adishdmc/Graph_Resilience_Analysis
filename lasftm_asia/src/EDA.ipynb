Importing Libraries

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

Load the Preprocessed LastFM Graph and Targets

```python
lastfm_graph = nx.read_gml('data/lastfm_graph.gml')
lastfm_targets = pd.read_csv('data/lastfm_targets_processed.csv')
```

Basic Information and Statistics

```python
print("Number of nodes:", lastfm_graph.number_of_nodes())
print("Number of edges:", lastfm_graph.number_of_edges())
print("Average degree:", sum(dict(lastfm_graph.degree()).values()) / lastfm_graph.number_of_nodes())
print("Density:", nx.density(lastfm_graph))
print("Average clustering coefficient:", nx.average_clustering(lastfm_graph))
```

Degree Distribution

```python
degrees = [degree for node, degree in lastfm_graph.degree()]
plt.figure(figsize=(10, 6))
plt.hist(degrees, bins=20, edgecolor='black')
plt.xlabel('Degree')
plt.ylabel('Frequency')
plt.title('Degree Distribution (LastFM Dataset)')
plt.show()
```

Clustering Coefficient Distribution

```python
clustering_coeffs = list(nx.clustering(lastfm_graph).values())
plt.figure(figsize=(10, 6))
plt.hist(clustering_coeffs, bins=20, edgecolor='black')
plt.xlabel('Clustering Coefficient')
plt.ylabel('Frequency')
plt.title('Clustering Coefficient Distribution (LastFM Dataset)')
plt.show()
```

Degree vs. Clustering Coefficient Scatterplot

```python
plt.figure(figsize=(10, 6))
plt.scatter(degrees, clustering_coeffs, alpha=0.6)
plt.xlabel('Degree')
plt.ylabel('Clustering Coefficient')
plt.title('Degree vs. Clustering Coefficient (LastFM Dataset)')
plt.show()
```

Betweenness Centrality Distribution

```python
betweenness = nx.betweenness_centrality(lastfm_graph)
betweenness_values = list(betweenness.values())
plt.figure(figsize=(10, 6))
plt.hist(betweenness_values, bins=20, edgecolor='black')
plt.xlabel('Betweenness Centrality')
plt.ylabel('Frequency')
plt.title('Betweenness Centrality Distribution (LastFM Dataset)')
plt.show()
```

Visualization of the Network

```python
plt.figure(figsize=(12, 12))
pos = nx.spring_layout(lastfm_graph, seed=42)
nx.draw(lastfm_graph, pos, with_labels=False, node_size=20, edge_color='k', alpha=0.5)
plt.title('LastFM Network Visualization')
plt.show()
```
